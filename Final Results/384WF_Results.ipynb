## Graph: Pending Time

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 8, 16, 32])

y111 = np.array([967, 370, 88, 67, 69])
y112 = np.array([416, 137, 37, 25, 25])
y113 = np.array([403, 99, 36, 22, 21])
y114 = np.array([3, 2, 1, 1, 1])

plt.plot(x, y111, 'o-', label="Slowest")
plt.plot(x, y112, 'o-', label="Mean")
plt.plot(x, y113, 'o-', label="Median")
plt.plot(x, y114, 'o-', label="Fastest")

plt.title("Pending Time vs Number of Nodes")
plt.xlabel('Number of Nodes')
plt.ylabel('Total time [s]')
plt.xticks(x)
plt.grid(True)

# Create a legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

## Speedup graph

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 8, 16, 32])  # Include 0 in the x values

y111 = np.array([967/967, 967/370, 967/88, 967/67, 967/69])
y112 = np.array([416/416, 416/137, 416/37, 416/25, 416/25])
y113 = np.array([403/403, 403/99, 403/36, 403/22, 403/21])
y114 = np.array([3/3, 3/2, 3/1, 3/1, 3/1])

plt.plot(x, y111, 'o-', label="Slowest")
plt.plot(x, y112, 'o-', label="Mean")
plt.plot(x, y113, 'o-', label="Median")
plt.plot(x, y114, 'o-', label="Fastest")

# Add the grey line
x_gray = np.array([2, 4, 8, 16, 37])
y_gray = x_gray / 2
plt.plot(x_gray, y_gray, '-', color='grey', label="Grey Line")

plt.title("Pending Speedup Time vs Number of Nodes")
plt.xlabel('Number of Nodes')
plt.ylabel('Speedup (w.r.t 2 nodes)')
plt.xticks(x)
plt.grid(True)

# Create a legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```
