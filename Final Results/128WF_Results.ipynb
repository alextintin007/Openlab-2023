# Graph: Total Time

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 8, 16, 32])  # Include 0 in the x values

y131 = np.array([367, 182, 94, 62, 57])
y132 = np.array([241, 117, 66, 50, 48])
y133 = np.array([262, 121, 67, 54, 50])
y134 = np.array([54, 39, 35, 34, 35])

plt.plot(x, y131, 'o-', label="Slowest")
plt.plot(x, y132, 'o-', label="Mean")
plt.plot(x, y133, 'o-', label="Median")
plt.plot(x, y134, 'o-', label="Fastest")

plt.title("Total Time vs Number of Nodes")
plt.xlabel('Number of Nodes')
plt.ylabel('Total time [s]')
plt.xticks(x)
plt.grid(True)

# Create a legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 8, 16, 32])  # Include 0 in the x values

y131 = np.array([367/367, 367/182, 367/94, 367/62, 367/57])
y132 = np.array([241/241, 241/117, 241/66, 241/50, 241/48])
y133 = np.array([262/262, 262/121, 262/67, 262/54, 262/50])
y134 = np.array([54/54, 54/39, 54/35, 54/34, 54/35])


plt.plot(x, y131, 'o-', label="Slowest")
plt.plot(x, y132, 'o-', label="Mean")
plt.plot(x, y133, 'o-', label="Median")
plt.plot(x, y134, 'o-', label="Fastest")

# Add the grey line
x_gray = np.array([2, 4, 16])
y_gray = np.array([367/367, 367/182, 8])
plt.plot(x_gray, y_gray, '--', color='grey', label="Ideal projection")

plt.title("Total Speedup Time vs Number of Nodes")
plt.xlabel('Number of Nodes')
plt.ylabel('Speedup (w.r.t 2 nodes)')
plt.xticks(x)
plt.grid(True)

# Create a legend outside the plot
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```
